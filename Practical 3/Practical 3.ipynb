```python
#IMPORTING LIBRARIES
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#LOADING DATASET
df = pd.read_csv("income.csv")
```

```python
df
```

```python
df.describe()
```

```python
#GROUPING AGE AND STORING IN A VARIBALE
grouped_data = df.groupby('Age')
```

```python
#PRINTING THE VARIABLE BY ITERATION
for age, group in grouped_data:
    print("Age:", age)
    print(group)
```

```python
#CALCULATING SUMMARY STATISTICS FOR THE 'INCOME' VARIABLE WITHIN EACH GROUP OF THE DATAFRAMEGROUPBY OBJECT GROUPED_DATA.
summary_stats = grouped_data.describe()
```
